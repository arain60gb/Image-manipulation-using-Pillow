# Rotating Images

```python
image = Image.open('demo_image.jpeg')

image_rot_90 = image.rotate(90)
image_rot_90.save('image_rot_90.jpg')

image_rot_180 = image.rotate(180)
image_rot_180.save('image_rot_180.jpg')
```

# Flipping Images

```python
image = Image.open('demo_image.jpeg')

image_flip = image.transpose(Image.FLIP_LEFT_RIGHT) # left_ right
image_flip.save('image_flip.jpg')
```

# Drawing on Images

```python
from PIL import Image, ImageDraw

canvas = Image.new('RGB', (400, 300), 'black')
img_draw = ImageDraw.Draw(canvas)
img_draw.rectangle((70, 50, 270, 200), outline='red', fill='green')
img_draw.text((200, 250), 'Hello World', fill='pink')
canvas.save('drawn_image.jpg')
```

```python
canvas
```

# Color Transforms

```python
image = Image.open('demo_image.jpeg')

greyscale_image = image.convert('L')
greyscale_image.save('greyscale_image.jpg')

print(image.mode) # Output: RGB
print(greyscale_image.mode) # Output: L
```

```python
greyscale_image
```

```python
image = Image.open('demo_image.jpeg')

red, green, blue = image.split()

print(image.mode) # Output: RGB
print(red.mode) # Output: L
print(green.mode) # Output: L
print(blue.mode) # Output: L

new_image = Image.merge("RGB", (green, red, blue))
new_image.save('new_image.jpg')

print(new_image.mode) # Output: RGB


# new image
```

```python
new_image
```

# Image enhancements

```python
from PIL import Image, ImageEnhance

image = Image.open('demo_image.jpeg')

contrast = ImageEnhance.Contrast(image)
contrast.enhance(0.5).save('contrast.jpg')
```

```python
contrast
```

```python
color = ImageEnhance.Color(image)
color.enhance(1.5).save('color.jpg') # this image save the name of color.jpg
```

```python
color
```

# Brightness

```python
brightness = ImageEnhance.Brightness(image)
brightness.enhance(0.5).save('brightness.jpg')
# increase the brightness
```

# Sharpness

```python
sharpness = ImageEnhance.Sharpness(image)
sharpness.enhance(0.5).save('sharpness.jpg')
# Sharpness is 10.5
```

# Resizing  Image

```python
image = Image.open('demo_image.jpeg')
new_image = image.resize((200, 200))
new_image.save('image_400.jpeg')

print(image.size) # Output: (1920, 1280)
print(new_image.size) # Output: (400, 400)
new_image
```

```python
image = Image.open('demo_image.jpeg')
image.thumbnail((100, 100))
image.save('image_thumbnail.jpeg')

print(image.size) # Output: (400, 267)
image
```
